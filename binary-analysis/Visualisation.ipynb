```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from glob import glob
```

```python
def transform(val):
    if not val:
        return 0
    if val[-1] == "}":
        val = val[:-1]
    if "{" in val:
        val = val.split("{")[1]
    return int(val.split(':')[1])
vtransform = pd.np.frompyfunc(transform, 1, 1)

def read_log(file):
    df = pd.read_csv(f, sep=",", header=None, skiprows=1, skipfooter=1)
    df.index = ["Binary", "Positions"]
    df[:] = vtransform(df.values[:,:])
    return df
```

```python
files = glob("./icon.pl.*.log")
```

# Options for residual calculations (and is the resulting residual noise?)

### Option 1: Difference 

### Option 2: XOR, Shifted, Shifted31

##### Temperature (easy), Wind (difficult)

# Difference (no gray)

```python
for f in [x for x in files if 'diff' in x and 'gray' not in x]:
    df = pd.read_csv(f, sep=",", header=None, skiprows=1, skipfooter=1)
    df.index = ["Binary", "Positions"]
    df[:] = vtransform(df.values[:,:])

    ax = (df.loc['Binary',:]/df.loc['Binary',1]*100).plot.bar(figsize=(15,7))
    plt.xticks(pd.np.arange(64), [bin(x)[2:] for x in pd.np.arange(64)]);
    plt.axhline(50, 0.03125, .0625);
    plt.axhline(25, .0625,.125);
    plt.axhline(12.5, .125, .25);
    plt.axhline(6.25, .25, .5);
    plt.axhline(3.125, .5, 1);
    plt.title(f);
    plt.axvline(1.5, color='gray', alpha=.5);
    plt.axvline(3.5, color='gray', alpha=.5);
    plt.axvline(7.5, color='gray', alpha=.5);
    plt.axvline(15.5, color='gray', alpha=.5);
    plt.axvline(31.5, color='gray', alpha=.5);
    plt.show()
```
